Below is an example of hoy to plot a graph using Jupyter notebook (Python language)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi, 100)

def sinusoidal_function(a,b,x):
    return a + b*np.sin(x)

f = sinusoidal_function(1, 2, x)

plt.figure(1)
plt.plot(x, f)
plt.title("Sinusoidal function")
plt.xlabel("x")
plt.xlabel("y")
plt.grid(True)
```
